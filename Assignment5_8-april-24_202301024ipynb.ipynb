```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
df=pd.read_csv('cluster_data.csv')
data=list(zip(df['Feature 1'],df['Feature 2']))
print(data)
kmeans=KMeans(n_clusters=2)
kmeans.fit(data)
plt.scatter(df['Feature 1'],df['Feature 2'],c=kmeans.labels_)
plt.show()
```
